### Работа спропусками данных
Ревизия наличия пропусков

```python
# вспомогательная функция для проверки и подсчета пропусков 
#(сразу isna() не догадалась использовать, оставлю на память, всё равно немного другой вид)
def count_na(data_t):
    col_nan = []
    col_nan_index = []
    for col in data_t:
        if len(data_t[data_t[col].isna()]) > 0:
            
            #print('Пропущенные значения в', data_t[col].name, len(data_t[data_t[col].isna()]))
            col_nan.append(len(data_t[data_t[col].isna()]))
            col_nan_index.append(data_t[col].name)
            col_nan_ser = pd.Series(data=col_nan, index=col_nan_index)
    col_nan_ser = col_nan_ser.sort_values(ascending=False)
    return (col_nan_ser)

display(count_na(data))

```

**Заполнение данных**

```python
#замена пропусков по статусу аппартаментов на значения False (квартиры)
data['is_apartment'] = data['is_apartment'].fillna(False) 
print('Значения столбца is_apartment теперь:', data['is_apartment'].unique())

#аналогичная замена по характеристике наличия балконов
#print(data['balcony'].unique()) #проверка значений и типа

data['balcony'] = data['balcony'].fillna(0.0) 
#тип данных был вещественный, при заполнении это был учтено, потом пустые данные были заменены на 0
data['balcony'] = data['balcony'].astype('int')
print('Значения столбца balcony теперь:', data['balcony'].unique())

```

#### Характеристика " Жилая площадь"


```python
#функция для первичной презентации данных, т.к. действия повторяются 
def analys_column(df,col):
    print(col.describe().round(1)) #округление для удобства чтения
    print('\n Гистограмма для характеристики', col.name)
    
    col= col.dropna() #для верной гистограммы уберем пустые значения
    
    plt.figure(figsize=(10,5))
    
    #гистограмма
    plt.subplot(1,2,1)
    col.hist(bins=100)
        
    #диаграмма размаха    
    plt.subplot(1, 2, 2)
    df.boxplot(column=col.name) 
    plt.show()
    
    #граница диапазона, выше которой много выбросов  
    normal_diapason_max = col.quantile(0.75) + 1.5*(col.quantile(0.75) - col.quantile(0.25))
    print(f'Верхняя граница "нормальных показателей" {normal_diapason_max:.1f}')
    
    outs_percent = len(df[df[col.name] > normal_diapason_max])/len(df)
    print(f'\n Процент выброса {outs_percent:.2%}')
   
    return

#функция для определения верхней границы, отсекающей выбросы
def max_norm(col):
    max_normal = col.quantile(0.75) + 1.5*(col.quantile(0.75) - col.quantile(0.25))
    return(max_normal)

analys_column(data,data['living_area'])


```

Для обработки недостоверных значений была взята следующая предпосылка:
*"Минимальный разрешенный размер студии по нормам СНиП - 19 м² (14 м² жилая площадь и 5 м² кухня)"* [ссылка](https://darstroy-yug.ru/articles/kvartira-studiya-vse-plyusy-i-minusy-etogo-zhilya/)

Таким образом, все строки с значением жилой площади менее 14м2 в дальнейшем срезе для анализа не используются.

С выбросами более "нормальных значений" на диаграме размаха, ситуация схожая с показателями общей площади, их также не рассматриваем в новом срезе.


```python
outs_living_area = max_norm(data['living_area'])

normal_living_area = data.query('living_area <= @outs_living_area').copy()
rich_living_area = data.query('living_area > @outs_living_area').copy()  #@outs_living_area

analys_column(normal_living_area, normal_living_area['living_area'])

print('Справочно: количество записей с превышением "нормальной жилой площади" (бизнес и элит класс)', rich_living_area.shape[0])
```

- на гистограмме видно 2 пика значений - до 20м2, 30м. Также есть чаще стречаемые значения в районе 40-45 м2.
Предположительно, **это связано с типами квартир**:
